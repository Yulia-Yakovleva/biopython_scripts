```python
### УРА ###

from Bio import SeqIO


def add_ebuchii_file(allSeqs, group_list, group, records_list = []):
    for element in group_list:
        for records in allSeqs:
            if str(records.name) == element:
                    records_list.append(records)
    print("Orthogroup_"+str(group)+"...")
    SeqIO.write(records_list, "orthogroup_"+str(group), "fasta")
    print("Done")
    

allSeqs = [seq_record for seq_record in SeqIO.parse('/home/yulia/streptococcus/faa_data/all_sequences.faa', 'fasta')]
for group, row in droped_ortho.iterrows():
    group_list = row.tolist()
    # print(len(group_list)) # 103, все нормуль
    add_ebuchii_file(allSeqs=allSeqs, group_list=group_list, group=group, records_list = [])
```
